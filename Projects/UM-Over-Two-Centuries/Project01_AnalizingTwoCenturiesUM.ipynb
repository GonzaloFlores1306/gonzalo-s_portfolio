```python
#Import libraries
import pandas as pd
import seaborn as sns
```

```python
#Import the data and show it
df = pd.read_csv("TWO_CENTURIES_OF_UM_RACES.csv")
df.head(10)
```

```python
#7 461 195 rows with 13 columns
df.shape
```

```python
#Watch the data type
df.dtypes
```

```python
#Clean up data
#Only want USA Races, 50km or 50Mi, 2020
```

```python
#Step 1: Show 50km or 50Mi
#Data with only 50km
df[df['Event distance/length'] == '50km']
```

```python
#Combine 50km in 50mi with isin
#Obtain the data with 50km or 50mi, since they're the same
filter_50 = df['Event distance/length'].isin(['50mi','50km'])
df[filter_50]
```

```python
#Obtain the data in the year 2020
filter_2020 = df['Year of event'] == 2020
df[filter_2020]
```

```python
#Obtain all the events in the USA

#Example: "wedqwdeqwdqwe (USA)"
#str.split('(') --> "wedqwdeqwdqwe (USA)" to ["wedqwdeqwdqwe","USA)"]
#str.get(1) --> "USA)"
#str.split(')') --> ["USA",""]
#str.get(0) --> "USA"
# == "USA" --> "USA" == "USA"
filter_USA = df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA'
df[filter_USA]
```

```python
#Combine all the filter together
combined_filters = filter_50 & filter_2020 & filter_USA
df2 = df[combined_filters]
df2
```

```python
#Remove USA from event name
#str.split('(') --> "wedqwdeqwdqwe (USA)" to ["wedqwdeqwdqwe","USA)"]
#str.get(0) --> "wedqwdeqwdqwe"
df2['Event name'].str.split('(').str.get(0)
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
df2.head()
```

```python
#Clean up athlete age (2020 - Athlete year of birth)
#Athlete age is a new column
df2['Athlete age'] = 2020 - df2['Athlete year of birth']
```

```python
#Remove h from the athlete performance
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5) #Watching the results from Athlete age and the modification of Athlete performance by eliminating 'h'
```

```python
#Drop unnecesary columns: Athlete Club, Athlete country, Athlete year of birth, Athlete Age Category 
df2 = df2.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'], axis=1)
df2.head()
```

```python
#Identify the columns with NaN values
df2.isna().sum()
```

```python
#Idenitfy rows with NabN values
df2[df2['Athlete age'].isna()==1]
```

```python
#Drop every row in the dataset with NaN values
df2 = df2.dropna()
df2.shape
```

```python
#Check for duplicated values
df2[df2.duplicated() == True]
```

```python
#reset index
df2.reset_index(drop = True)
```

```python
#fix types of the columns
df2.dtypes
```

```python
#Changing the data type of columns
df2['Athlete age'] = df2['Athlete age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#Rename columns 

#Year of event                  int64
#Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed         object
#Athlete ID                     int64
#Athlete age                  float64
#dtype: object
```

```python
df2 = df2.rename(columns = {
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id',
    'Athlete age': 'athlete_age'
})  
```

```python
#Seasons for the data -> Slower in summer than winter?

#Number of months for every season of the year
#spring 3-5
#summer  6-8
#fall 9-11
#winter 12-2

#split between two decimals. 
```

```python
#I'll use the race_day
df3.head()
```

```python
#New column month
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
df3.head()
```

```python
#Applying the season's logic with lambda function. New colum season
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if  x>11 else 'Fall' if x>8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
df3
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
#In Summer, they ran 1km slower than in Spring
```
